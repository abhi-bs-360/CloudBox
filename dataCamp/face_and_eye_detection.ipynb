```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
#%matplotlib inline

print('Done')
```

```python
face_class = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
#eye_class = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
```

```python
img = cv2.imread('photo/face.jpg')
fixed = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(fixed)
```

```python
faces = face_class.detectMultiScale(fixed, 1.3, 5)
```

```python
def detect(fixed):
    face_rects = face_class.detectMultiScale(fixed)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(fixed, (x,y), (x+w,y+w), (255,0,0),7)
        
        '''eye_rects = eye_class.detectMultiScale(fixed)
        for (ix,iy,iw,ih) in eye_rects:
            cv2.rectangle(fixed, (ix,iy), (ix+iw,iy+ih), (255,255,255), 5)'''
        
    return fixed
```

```python
res = detect(fixed)
plt.imshow(res)
```
